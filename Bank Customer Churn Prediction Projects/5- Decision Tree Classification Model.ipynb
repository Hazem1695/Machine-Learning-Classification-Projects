# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

# **Outlier Analysis**
Outlier detection was performed using the IQR method on the numerical features.

### Features with Detected Outliers

- `CreditScore`
- `Age`
- `NumOfProducts`
- `Exited`

### Decision About Outliers

The detected outliers were intentionally retained instead of being removed because these values represent realistic customer behavior rather than data entry errors.

#### CreditScore
Extreme credit scores are valid in banking datasets and may contain important information about customer financial behavior and churn patterns. Removing these observations could reduce the model’s ability to learn from high-risk or low-risk customers.

#### Age
Older customers were detected as statistical outliers due to the IQR method. However, these customers are valid observations and may have significantly different churn behavior compared to younger customers. Therefore, they were preserved.

#### NumOfProducts
Customers with a high number of banking products were identified as outliers. These cases are meaningful from a business perspective because product usage can strongly influence customer retention and churn.

#### Exited
The `Exited` feature represents the target variable of the classification problem, where:

- `0` = Customer stayed
- `1` = Customer churned

Although the IQR method detected many values as outliers, this result is not meaningful because `Exited` is a binary categorical variable rather than a continuous numerical feature.

Outlier detection methods such as IQR are designed for continuous numerical distributions and should not be applied to classification labels.

Therefore, no outlier handling was performed on the `Exited` column, and all observations were retained.

### Additional Notes
- Identifier columns such as:
  - `RowNumber`
  - `CustomerId`

  were removed from the dataset because they do not provide predictive value.

- `Surname`

The `Surname` column was removed from the dataset because it represents a high cardinality categorical feature with a large number of unique customer names.

Customer surnames are unlikely to have a direct causal relationship with churn behavior and may introduce noise into the model.

Additionally, encoding this feature could create unnecessary dimensionality or artificial relationships between categories without providing meaningful predictive value.

Therefore, the `Surname` feature was excluded from the modeling process.

### Conclusion

The outliers were retained to preserve potentially valuable business information and avoid removing legitimate customer behavior patterns. Tree-based models such as Random Forest are generally robust to outliers, making aggressive outlier removal unnecessary for this project.

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False, drop='first'),[1,2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
```

### Categorical Feature Encoding

Categorical variables were transformed into numerical representations using One-Hot Encoding.

#### Geography
The `Geography` feature was encoded using One-Hot Encoding. Since the feature contains three categories (`France`, `Spain`, and `Germany`), two binary columns were created while one category was automatically dropped to avoid the dummy variable trap and multicollinearity.

Encoding representation:

| Germany | Spain | Geography |
|---|---|---|
| 0 | 0 | France |
| 0 | 1 | Spain |
| 1 | 0 | Germany |

#### Gender
The `Gender` feature was also transformed using One-Hot Encoding. Since it contains only two categories (`Female` and `Male`), a single binary column was generated after dropping one category.

Encoding representation:

| Gender |
|---|
| 0 → Female |
| 1 → Male |

These transformations convert categorical variables into numerical representations suitable for machine learning algorithms.

# **Building Decision Tree Classification Model**

## Training the Building Decision Tree Classification model on the Training set

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(max_depth=6, min_samples_split=3, min_samples_leaf=20, criterion='entropy', class_weight = 'balanced', random_state=0)
classifier.fit(X_train, y_train)

#{'criterion': 'entropy', 'max_depth': 6, 'min_samples_leaf': 3, 'min_samples_split': 20} from GridSearchCV
```

# **Preticting the Test set results**

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)[0:100,:])
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)

print("\nMean Accuracy:")
print(accuracies.mean())

print("\nStandard Deviation:")
print(accuracies.std())
```

# Decision Tree Classifier Model Comparison and Evaluation

This experiment focused on evaluating different configurations of Decision Tree models for an imbalanced classification problem. The comparison investigated the impact of:

* Balanced class weights
* Hyperparameter tuning
* Feature scaling strategies
* Tree depth optimization
* Minority-class detection capability
* Model generalization and stability

The experiments aimed to determine which Decision Tree configuration provides the strongest balance between predictive performance and minority-class detection.

---

# 1. Decision Tree Classifier Using StandardScaler Without Balanced Class Weights

The baseline Decision Tree model achieved an accuracy score of `79.56%` and demonstrated moderate performance on the minority class.

The model achieved:

* Minority precision = `0.50`
* Minority recall = `0.54`
* Minority F1-score = `0.52`
* Receiver Operating Characteristic - Area Under Curve = `0.699`

Compared to previous experiments with other algorithms, this baseline Decision Tree model already demonstrated significantly stronger minority-class detection capability without requiring imbalance-handling techniques such as Synthetic Minority Oversampling Technique.

The model maintained a relatively balanced trade-off between majority-class accuracy and minority-class detection.

This experiment highlights one of the advantages of Decision Trees: they naturally handle some imbalance patterns better than probabilistic and linear models.

---

# 2. Decision Tree Classifier Using StandardScaler With Balanced Class Weights

After introducing balanced class weights, the model performance improved across nearly all evaluation metrics.

The model achieved:

* Accuracy score = `81.40%`
* Minority recall = `0.56`
* Minority F1-score = `0.55`
* Receiver Operating Characteristic - Area Under Curve = `0.719`

This configuration improved:

* Minority-class recall
* Minority-class F1-score
* Receiver Operating Characteristic - Area Under Curve
* Overall accuracy

The results demonstrate that balanced class weights successfully encouraged the model to pay more attention to minority-class samples without significantly damaging majority-class performance.

This experiment confirms the effectiveness of class-weight balancing in Decision Tree classifiers.

---

# 3. Decision Tree Classifier Using RobustScaler With Balanced Class Weights

Using RobustScaler instead of StandardScaler produced nearly identical results.

The model achieved:

* Accuracy score = `81.32%`
* Minority recall = `0.56`
* Minority F1-score = `0.55`
* Receiver Operating Characteristic - Area Under Curve = `0.718`

The minimal difference between StandardScaler and RobustScaler indicates that Decision Trees are relatively insensitive to feature scaling because they rely on feature splits rather than distance-based calculations.

This experiment demonstrates an important machine learning concept:

* scaling is generally less critical for tree-based algorithms compared to distance-based models such as Support Vector Machines or K-Nearest Neighbors.

---

# 4. Tuned Decision Tree Classifier With Maximum Depth = 6 Using RobustScaler

After hyperparameter tuning, the Decision Tree model demonstrated a substantial improvement in minority-class detection capability.

The model achieved:

* Accuracy score = `75.72%`
* Minority recall = `0.80`
* Minority F1-score = `0.57`
* Receiver Operating Characteristic - Area Under Curve = `0.772`

This configuration produced one of the strongest recall scores among all Decision Tree experiments.

Although overall accuracy decreased compared to earlier models, the model became significantly better at identifying minority-class samples.

The improvement in Receiver Operating Characteristic - Area Under Curve indicates stronger class separation capability and improved overall predictive balance.

This experiment demonstrates the importance of controlling tree complexity and tuning hyperparameters in Decision Tree classifiers.

---

# 5. Tuned Decision Tree Classifier With Maximum Depth = 7 Using RobustScaler

Increasing the maximum depth to `7` produced slightly improved overall performance.

The model achieved:

* Accuracy score = `76.12%`
* Minority recall = `0.79`
* Minority F1-score = `0.57`
* Receiver Operating Characteristic - Area Under Curve = `0.773`

Compared to the previous configuration, this model maintained strong minority-class detection while slightly improving overall stability and predictive balance.

The Receiver Operating Characteristic - Area Under Curve score became one of the highest among all Decision Tree experiments.

This experiment demonstrates that moderate increases in tree depth can improve the model’s ability to capture more complex decision boundaries.

---

# 6. Tuned Decision Tree Classifier With Maximum Depth = 7 Using StandardScaler

Using StandardScaler instead of RobustScaler produced identical results.

The model achieved:

* Accuracy score = `76.12%`
* Minority recall = `0.79`
* Minority F1-score = `0.57`
* Receiver Operating Characteristic - Area Under Curve = `0.773`

This result further confirms that feature scaling has minimal influence on Decision Tree performance.

The identical metrics strengthen the conclusion that tree-based models are largely independent of scaling techniques.

---

# 7. Best Hyperparameter-Tuned Decision Tree Classifier From Grid Search

The final optimized Decision Tree model, selected using grid search, produced the strongest overall Decision Tree performance.

The model achieved:

* Accuracy score = `76.92%`
* Minority recall = `0.79`
* Minority F1-score = `0.58`
* Receiver Operating Characteristic - Area Under Curve = `0.778`

This configuration provided:

* The highest Receiver Operating Characteristic - Area Under Curve score
* The best minority-class F1-score
* Strong recall performance
* Improved overall predictive balance

The hyperparameter optimization successfully controlled overfitting while maintaining strong sensitivity toward minority-class samples.

This experiment demonstrates the effectiveness of systematic hyperparameter tuning in improving Decision Tree performance for imbalanced classification problems.
