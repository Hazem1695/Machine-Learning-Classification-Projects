# Final Machine Learning Experimentation Documentation

# Project Overview

This project presents a comprehensive machine learning experimentation workflow for solving a customer churn prediction problem using supervised classification algorithms.

The primary objective of the project was to analyze and compare multiple machine learning models under different preprocessing, imbalance-handling, and hyperparameter tuning strategies in order to identify the most effective model for predicting customer churn.

The project focused heavily on:

* Imbalanced classification handling
* Comparative model analysis
* Hyperparameter optimization
* Model evaluation using business-oriented metrics
* Generalization capability
* Minority class detection performance

Each machine learning model was implemented in a separate notebook containing all experiments related to that algorithm, including preprocessing variations, scaling strategies, balancing techniques, and parameter tuning.

The final goal was not only to achieve high accuracy, but to build a reliable churn prediction system capable of effectively identifying minority class samples while maintaining strong generalization performance.

---

# Machine Learning Models Evaluated

The following machine learning algorithms were implemented and evaluated throughout the project:

1. Logistic Regression
2. K-Nearest Neighbors
3. Support Vector Classification
4. Gaussian Naive Bayes
5. Decision Tree Classifier
6. Random Forest Classifier

---

# Experimentation Methodology

The experimentation workflow followed a structured machine learning pipeline:

1. Data preprocessing
2. Encoding categorical variables
3. Feature scaling
4. Handling imbalanced data
5. Model training
6. Hyperparameter tuning
7. Cross-validation
8. Performance evaluation
9. Comparative analysis

The experiments included:

* StandardScaler
* RobustScaler
* Synthetic Minority Oversampling Technique
* Balanced class weights
* Hyperparameter optimization
* Cross-validation analysis

The evaluation process focused primarily on:

* Minority class recall
* F1-Score
* Receiver Operating Characteristic - Area Under Curve
* Model stability
* Generalization performance

Accuracy alone was not considered sufficient because the dataset was imbalanced.

---

# Logistic Regression Experiments

Logistic Regression served as the baseline model for the project.

The initial experiments without imbalance handling achieved relatively high accuracy but failed to identify minority class samples effectively. The model showed strong bias toward the majority class and produced very low recall for churn prediction.

Applying `class_weight='balanced'` significantly improved the model’s ability to detect churned customers. Minority recall improved dramatically, and Receiver Operating Characteristic - Area Under Curve performance increased considerably.

Both StandardScaler and RobustScaler produced nearly identical results, demonstrating the stability of Logistic Regression across scaling methods.

### Main Findings

* Stable and interpretable model
* Strong baseline performance
* Significant improvement after imbalance handling
* Limited nonlinear learning capability
* Good generalization performance

---

# K-Nearest Neighbors Experiments

K-Nearest Neighbors demonstrated moderate performance across the experiments.

Without imbalance handling, the model achieved good accuracy but struggled with minority class detection. Applying Synthetic Minority Oversampling Technique after feature scaling improved recall and Receiver Operating Characteristic - Area Under Curve performance significantly.

One of the most important findings was that preprocessing order strongly affected performance. Scaling before Synthetic Minority Oversampling Technique produced substantially better results than oversampling before scaling.

The experiments confirmed that K-Nearest Neighbors is highly sensitive to feature scaling because it relies on distance-based learning.

### Main Findings

* Highly dependent on feature scaling
* StandardScaler outperformed RobustScaler
* Sensitive to preprocessing order
* Synthetic Minority Oversampling Technique improved minority detection
* Less stable than ensemble and margin-based algorithms

---

# Support Vector Classification Experiments

Support Vector Classification was one of the strongest performing algorithms in the project.

Several kernels were evaluated:

* Linear
* Radial Basis Function
* Polynomial
* Sigmoid

The linear kernel failed under imbalance conditions when balancing techniques were not applied. After enabling balanced class weights, minority class detection improved significantly.

The Radial Basis Function kernel achieved the strongest Support Vector Classification performance and demonstrated excellent nonlinear separation capability and strong generalization performance.

Polynomial kernels also performed well, while sigmoid kernels produced weak results and were unsuitable for this dataset.

### Main Findings

* Radial Basis Function kernel achieved the best Support Vector Classification performance
* Strong nonlinear learning capability
* Excellent Receiver Operating Characteristic - Area Under Curve performance
* Strong minority class recall
* Strong generalization performance

---

# Gaussian Naive Bayes Experiments

Gaussian Naive Bayes produced the weakest overall performance among all evaluated algorithms.

The baseline model failed completely to detect minority class samples because of dataset imbalance and the strong feature independence assumptions of Gaussian Naive Bayes.

Applying manual priors and Synthetic Minority Oversampling Technique improved recall but introduced instability and excessive false positives.

Performance improved after combining encoding techniques with imbalance handling strategies, but the model still underperformed compared to more advanced algorithms.

These experiments demonstrated that Gaussian Naive Bayes was not suitable for this dataset because the feature relationships violated the algorithm’s assumptions.

### Main Findings

* Weak baseline performance
* Strong dependence on feature assumptions
* Sensitive to imbalance
* Improved after encoding and balancing
* Not suitable for final deployment

---

# Decision Tree Experiments

Decision Tree models achieved strong classification performance and excellent minority class recall.

Initial untuned trees achieved moderate performance, but hyperparameter tuning significantly improved Receiver Operating Characteristic - Area Under Curve performance and reduced overfitting risk.

The experiments demonstrated that controlling:

* Maximum depth
* Minimum samples split
* Minimum samples leaf

greatly improved generalization capability.

Unlike distance-based algorithms, Decision Trees were almost unaffected by feature scaling methods.

The tuned Decision Tree achieved one of the strongest recall scores among all evaluated models.

### Main Findings

* Strong minority class recall
* Hyperparameter tuning significantly improved performance
* Excellent interpretability
* Higher overfitting risk than ensemble models
* Scaling had minimal impact

---

# Random Forest Experiments

Random Forest achieved the strongest overall performance among all evaluated models.

The baseline model produced very high accuracy, but tuning the model significantly improved minority class detection and overall balance between recall and Receiver Operating Characteristic - Area Under Curve performance.

The optimized Random Forest model achieved the highest Receiver Operating Characteristic - Area Under Curve score in the project while maintaining strong recall and stable cross-validation performance.

The experiments demonstrated the effectiveness of ensemble learning for imbalanced classification problems.

Random Forest successfully balanced:

* Generalization capability
* Minority class detection
* Stability
* Overfitting resistance

### Main Findings

* Best overall model
* Highest Receiver Operating Characteristic - Area Under Curve score
* Strong balance between recall and precision
* Strong generalization capability
* Stable performance

# Major Insights From the Experiments

## Imbalanced Classification Handling Was Critical

The experiments clearly demonstrated that imbalance handling dramatically affected model performance.

Without balancing techniques:

* Several models completely failed to detect minority class samples
* Accuracy appeared misleadingly high
* Minority recall collapsed

Applying:

* Balanced class weights
* Synthetic Minority Oversampling Technique

significantly improved minority class detection.

---

## Accuracy Alone Was Misleading

The experiments confirmed that high accuracy does not necessarily indicate good classification performance in imbalanced datasets.

Several models achieved high accuracy while failing to detect churned customers effectively.

Receiver Operating Characteristic - Area Under Curve, recall, and F1-Score provided more meaningful evaluation metrics.

---

## Feature Scaling Strongly Affected Distance-Based Models

The experiments demonstrated clear differences between algorithm families.

### Strongly affected by scaling:

* K-Nearest Neighbors
* Support Vector Classification
* Logistic Regression

### Minimally affected by scaling:

* Decision Tree
* Random Forest

---

## Preprocessing Order Was Important

Applying feature scaling before Synthetic Minority Oversampling Technique produced significantly better results for distance-based models.

Incorrect preprocessing order negatively affected:

* K-Nearest Neighbors
* Gaussian Naive Bayes

---

## Hyperparameter Tuning Produced Major Improvements

Hyperparameter optimization significantly improved:

* Decision Tree performance
* Random Forest generalization
* Support Vector Classification performance

Tuning improved:

* Minority class detection
* Stability
* Receiver Operating Characteristic - Area Under Curve performance
* Overfitting control

---

## Ensemble Learning Produced the Best Overall Results

The experiments demonstrated that ensemble learning methods achieved stronger overall performance than standalone algorithms.

Random Forest achieved:

* The highest Receiver Operating Characteristic - Area Under Curve score
* Strong generalization capability
* Strong stability
* Effective minority class detection

---

# Final Conclusion

This project provided a comprehensive practical study of machine learning experimentation for customer churn prediction under imbalanced classification conditions.

The experiments demonstrated that:

* Different algorithms react differently to imbalance handling techniques
* Proper preprocessing significantly affects performance
* Hyperparameter tuning is essential for strong generalization
* Ensemble learning methods outperform simpler models in complex classification tasks

Among all evaluated algorithms, the optimized Random Forest model achieved the strongest overall performance and was selected as the final recommended deployment model.

The final Random Forest model successfully balanced:

* Receiver Operating Characteristic - Area Under Curve performance
* Minority class recall
* Stability
* Generalization capability

This project represents a complete end-to-end machine learning experimentation workflow focused on real-world classification analysis, comparative evaluation, and model optimization.
