# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

# **Outlier Analysis**
Outlier detection was performed using the IQR method on the numerical features.

### Features with Detected Outliers

- `CreditScore`
- `Age`
- `NumOfProducts`
- `Exited`

### Decision About Outliers

The detected outliers were intentionally retained instead of being removed because these values represent realistic customer behavior rather than data entry errors.

#### CreditScore
Extreme credit scores are valid in banking datasets and may contain important information about customer financial behavior and churn patterns. Removing these observations could reduce the model’s ability to learn from high-risk or low-risk customers.

#### Age
Older customers were detected as statistical outliers due to the IQR method. However, these customers are valid observations and may have significantly different churn behavior compared to younger customers. Therefore, they were preserved.

#### NumOfProducts
Customers with a high number of banking products were identified as outliers. These cases are meaningful from a business perspective because product usage can strongly influence customer retention and churn.

#### Exited
The `Exited` feature represents the target variable of the classification problem, where:

- `0` = Customer stayed
- `1` = Customer churned

Although the IQR method detected many values as outliers, this result is not meaningful because `Exited` is a binary categorical variable rather than a continuous numerical feature.

Outlier detection methods such as IQR are designed for continuous numerical distributions and should not be applied to classification labels.

Therefore, no outlier handling was performed on the `Exited` column, and all observations were retained.

### Additional Notes
- Identifier columns such as:
  - `RowNumber`
  - `CustomerId`

  were removed from the dataset because they do not provide predictive value.

- `Surname`

The `Surname` column was removed from the dataset because it represents a high cardinality categorical feature with a large number of unique customer names.

Customer surnames are unlikely to have a direct causal relationship with churn behavior and may introduce noise into the model.

Additionally, encoding this feature could create unnecessary dimensionality or artificial relationships between categories without providing meaningful predictive value.

Therefore, the `Surname` feature was excluded from the modeling process.

### Conclusion

The outliers were retained to preserve potentially valuable business information and avoid removing legitimate customer behavior patterns. Tree-based models such as Random Forest are generally robust to outliers, making aggressive outlier removal unnecessary for this project.

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False, drop='first'),[1,2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
```

### Categorical Feature Encoding

Categorical variables were transformed into numerical representations using One-Hot Encoding.

#### Geography
The `Geography` feature was encoded using One-Hot Encoding. Since the feature contains three categories (`France`, `Spain`, and `Germany`), two binary columns were created while one category was automatically dropped to avoid the dummy variable trap and multicollinearity.

Encoding representation:

| Germany | Spain | Geography |
|---|---|---|
| 0 | 0 | France |
| 0 | 1 | Spain |
| 1 | 0 | Germany |

#### Gender
The `Gender` feature was also transformed using One-Hot Encoding. Since it contains only two categories (`Female` and `Male`), a single binary column was generated after dropping one category.

Encoding representation:

| Gender |
|---|
| 0 → Female |
| 1 → Male |

These transformations convert categorical variables into numerical representations suitable for machine learning algorithms.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)

X_train_smote, y_train_smote = smote.fit_resample(X_train,y_train)
```

# **Building KNearest Neighbors (KNN) Classification Model**

## Training the Building KNearest Neighbors (KNN) Classification model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
classifier.fit(X_train_smote, y_train_smote)
```

# **Preticting the Test set results**

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)[0:100,:])
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)

print("\nMean Accuracy:")
print(accuracies.mean())

print("\nStandard Deviation:")
print(accuracies.std())
```

# Model Comparison and Evaluation

This experiment focused on evaluating the impact of different preprocessing strategies on an imbalanced classification problem. The comparison analyzed the effect of:

* Applying Synthetic Minority Oversampling Technique
* Changing the preprocessing order
* Using different feature scaling techniques
* Evaluating model stability and minority-class detection performance

The experiments were designed to investigate how preprocessing decisions influence classification performance beyond overall accuracy.

---

# 1. Baseline Model — Without Synthetic Minority Oversampling Technique Using StandardScaler

The baseline model achieved the highest overall accuracy score of `83.04%`, indicating strong general performance on the dataset. The model also achieved the highest precision score for the minority class (`0.61`), meaning that when the model predicted the positive class, it was more likely to be correct compared to other experiments.

However, despite the strong accuracy, the recall score for the minority class was only `0.48`. This means the model failed to detect more than half of the minority-class samples. In imbalanced classification problems, relying only on accuracy can be misleading because the model may favor the majority class while underperforming on the minority class.

The Receiver Operating Characteristic - Area Under Curve score was `0.698`, showing moderate class discrimination capability. This baseline experiment demonstrates that strong accuracy does not necessarily indicate effective minority-class detection.

---

# 2. Applying Synthetic Minority Oversampling Technique After Scaling Using StandardScaler

This experiment produced the strongest overall balance between minority-class detection and general model performance.

After applying Synthetic Minority Oversampling Technique after feature scaling, the recall score for the minority class increased significantly from `0.48` to `0.67`. This improvement shows that the model became substantially better at identifying minority-class instances.

The Receiver Operating Characteristic - Area Under Curve score also increased to `0.723`, which was the highest among all experiments. This indicates stronger class separation and improved predictive capability.

Although overall accuracy decreased from `83.04%` to `75.52%`, this decrease is expected in imbalanced classification tasks because the model becomes less biased toward the majority class and more sensitive to minority-class detection.

This experiment demonstrates the importance of applying feature scaling before Synthetic Minority Oversampling Technique, since Synthetic Minority Oversampling Technique relies on distance-based nearest-neighbor calculations.

---

# 3. Applying Synthetic Minority Oversampling Technique After Scaling Using RobustScaler

Using RobustScaler after applying Synthetic Minority Oversampling Technique also produced strong results.

The recall score for the minority class reached `0.64`, while the Receiver Operating Characteristic - Area Under Curve score reached `0.708`. The performance remained competitive and close to the StandardScaler experiment.

This experiment demonstrates that RobustScaler can perform effectively in datasets containing outliers or skewed numerical distributions because it scales features using the median and interquartile range instead of mean and standard deviation.

Although the performance was slightly lower than StandardScaler, the experiment highlights the importance of testing multiple preprocessing strategies rather than relying on a single approach.

---

# 4. Applying Synthetic Minority Oversampling Technique Before Scaling Using StandardScaler

This experiment produced the weakest performance among all experiments.

The recall score for the minority class dropped dramatically to `0.19`, while the Receiver Operating Characteristic - Area Under Curve score decreased to `0.565`.

This result demonstrates that applying Synthetic Minority Oversampling Technique before feature scaling negatively impacts the quality of synthetic samples. Since Synthetic Minority Oversampling Technique depends on nearest-neighbor distance calculations, unscaled features distort the distance measurements and reduce the effectiveness of oversampling.

Although the overall accuracy remained relatively high at `78.72%`, the model performed poorly in detecting minority-class samples.

This experiment strongly illustrates the importance of preprocessing order in machine learning pipelines.

---

# 5. Applying Synthetic Minority Oversampling Technique Before Scaling Using RobustScaler

This experiment achieved moderate performance compared to the other configurations.

The recall score for the minority class reached `0.63`, while the Receiver Operating Characteristic - Area Under Curve score reached `0.711`.

Although the preprocessing order was not scientifically optimal, the model still achieved relatively balanced performance. However, the results remained slightly weaker than applying Synthetic Minority Oversampling Technique after scaling.

This experiment reinforces the importance of preprocessing order while also showing that RobustScaler can partially reduce the negative impact caused by improper scaling order.

# Key Insights

## Accuracy Alone Is Misleading in Imbalanced Classification

The baseline model achieved the highest accuracy score but failed to detect a large portion of minority-class samples. This demonstrates that accuracy alone is not a reliable evaluation metric for imbalanced datasets.

---

## Synthetic Minority Oversampling Technique Improved Minority-Class Detection

Applying Synthetic Minority Oversampling Technique significantly increased minority-class recall across most experiments. This confirms the effectiveness of oversampling techniques in improving class balance during training.

---

## Preprocessing Order Strongly Influences Model Performance

The experiments clearly showed that applying feature scaling before Synthetic Minority Oversampling Technique produces better results. Incorrect preprocessing order negatively affected recall, F1-score, and Receiver Operating Characteristic - Area Under Curve performance.

---

## StandardScaler Produced the Best Overall Results

Among all preprocessing strategies, StandardScaler combined with applying Synthetic Minority Oversampling Technique after scaling produced the strongest balance between recall and Receiver Operating Characteristic - Area Under Curve performance.

---

## RobustScaler Demonstrated Strong Stability

RobustScaler achieved competitive results and showed strong stability in handling numerical features that may contain outliers or skewed distributions.

---

# Conclusion

This experimentation demonstrated that preprocessing strategy selection significantly impacts classification performance in imbalanced machine learning problems.

The results showed that:

* Applying Synthetic Minority Oversampling Technique improved minority-class detection capability
* Feature scaling should be applied before Synthetic Minority Oversampling Technique
* Accuracy alone should not be used as the primary evaluation metric in imbalanced datasets
* Receiver Operating Characteristic - Area Under Curve, recall, and F1-score provide more reliable performance evaluation

The strongest overall configuration was applying Synthetic Minority Oversampling Technique after scaling using StandardScaler, as it achieved the best balance between minority-class detection and overall predictive capability.

These experiments highlight the importance of systematic preprocessing experimentation and careful evaluation when building reliable machine learning classification pipelines.
