# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

# **Outlier Analysis**
Outlier detection was performed using the IQR method on the numerical features.

### Features with Detected Outliers

- `CreditScore`
- `Age`
- `NumOfProducts`
- `Exited`

### Decision About Outliers

The detected outliers were intentionally retained instead of being removed because these values represent realistic customer behavior rather than data entry errors.

#### CreditScore
Extreme credit scores are valid in banking datasets and may contain important information about customer financial behavior and churn patterns. Removing these observations could reduce the model’s ability to learn from high-risk or low-risk customers.

#### Age
Older customers were detected as statistical outliers due to the IQR method. However, these customers are valid observations and may have significantly different churn behavior compared to younger customers. Therefore, they were preserved.

#### NumOfProducts
Customers with a high number of banking products were identified as outliers. These cases are meaningful from a business perspective because product usage can strongly influence customer retention and churn.

#### Exited
The `Exited` feature represents the target variable of the classification problem, where:

- `0` = Customer stayed
- `1` = Customer churned

Although the IQR method detected many values as outliers, this result is not meaningful because `Exited` is a binary categorical variable rather than a continuous numerical feature.

Outlier detection methods such as IQR are designed for continuous numerical distributions and should not be applied to classification labels.

Therefore, no outlier handling was performed on the `Exited` column, and all observations were retained.

### Additional Notes
- Identifier columns such as:
  - `RowNumber`
  - `CustomerId`

  were removed from the dataset because they do not provide predictive value.

- `Surname`

The `Surname` column was removed from the dataset because it represents a high cardinality categorical feature with a large number of unique customer names.

Customer surnames are unlikely to have a direct causal relationship with churn behavior and may introduce noise into the model.

Additionally, encoding this feature could create unnecessary dimensionality or artificial relationships between categories without providing meaningful predictive value.

Therefore, the `Surname` feature was excluded from the modeling process.

### Conclusion

The outliers were retained to preserve potentially valuable business information and avoid removing legitimate customer behavior patterns. Tree-based models such as Random Forest are generally robust to outliers, making aggressive outlier removal unnecessary for this project.

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False, drop='first'),[1,2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
#
```

### Categorical Feature Encoding

Categorical variables were transformed into numerical representations using One-Hot Encoding.

#### Geography
The `Geography` feature was encoded using One-Hot Encoding. Since the feature contains three categories (`France`, `Spain`, and `Germany`), two binary columns were created while one category was automatically dropped to avoid the dummy variable trap and multicollinearity.

Encoding representation:

| Germany | Spain | Geography |
|---|---|---|
| 0 | 0 | France |
| 0 | 1 | Spain |
| 1 | 0 | Germany |

#### Gender
The `Gender` feature was also transformed using One-Hot Encoding. Since it contains only two categories (`Female` and `Male`), a single binary column was generated after dropping one category.

Encoding representation:

| Gender |
|---|
| 0 → Female |
| 1 → Male |

These transformations convert categorical variables into numerical representations suitable for machine learning algorithms.

# **Building Naive Bayes Classification Model**

## Training the Building Naive Bayes Classification model on the Training set

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train_smote, y_train_smote)
```

# **Preticting the Test set results**

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)[0:100,:])
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)

print("\nMean Accuracy:")
print(accuracies.mean())

print("\nStandard Deviation:")
print(accuracies.std())
```

# Gaussian Naive Bayes Model Comparison and Evaluation

This experiment focused on evaluating different configurations of Gaussian Naive Bayes models for an imbalanced classification problem. The comparison investigated the impact of:

* Synthetic Minority Oversampling Technique
* Prior probability adjustment
* Feature scaling strategies
* One-hot encoding strategies
* Minority-class detection capability
* Model stability and predictive balance

The experiments aimed to determine how preprocessing and imbalance-handling strategies influence the performance of Gaussian Naive Bayes classifiers.

---

# 1. Gaussian Naive Bayes Without Synthetic Minority Oversampling Technique Using StandardScaler

The baseline Gaussian Naive Bayes model achieved an overall accuracy score of `79.64%`. However, the confusion matrix revealed that the model predicted every sample as the majority class and completely failed to classify minority-class samples.

As a result:

* Minority-class precision = `0.00`
* Minority-class recall = `0.00`
* Minority-class F1-score = `0.00`
* Receiver Operating Characteristic - Area Under Curve = `0.50`

A Receiver Operating Characteristic - Area Under Curve score of `0.50` indicates no effective class separation capability beyond random guessing.

The warning messages generated during evaluation confirmed that the model never predicted minority-class labels.

This experiment demonstrates that Gaussian Naive Bayes struggles significantly with imbalanced datasets when no imbalance-handling strategy is applied.

---

# 2. Gaussian Naive Bayes With Balanced Priors Using StandardScaler

After adjusting the class prior probabilities to `[0.5, 0.5]`, the model demonstrated a major improvement in minority-class detection.

The minority-class recall increased dramatically to `0.68`, while the Receiver Operating Characteristic - Area Under Curve score improved to `0.623`.

This improvement indicates that manually balancing class priors encouraged the model to become more sensitive to minority-class samples.

However, this configuration also produced:

* Lower overall accuracy = `58.76%`
* High number of false positives
* Reduced majority-class performance

Although the model became more balanced, the trade-off between recall and precision remained significant.

This experiment highlights the importance of prior probability adjustment in probabilistic classifiers when dealing with imbalanced datasets.

---

# 3. Gaussian Naive Bayes With Synthetic Minority Oversampling Technique Using StandardScaler

Applying Synthetic Minority Oversampling Technique improved minority-class recall further to `0.76`, which was one of the highest recall values across all Gaussian Naive Bayes experiments.

However, the model suffered from severe performance imbalance:

* Accuracy dropped to `48.80%`
* Minority-class precision remained very low at `0.25`
* Receiver Operating Characteristic - Area Under Curve remained relatively weak at `0.589`

The confusion matrix showed a very large number of false positives, meaning the model over-predicted the minority class.

This experiment demonstrates that while Synthetic Minority Oversampling Technique improved sensitivity toward minority samples, it also introduced instability and excessive false-positive predictions in Gaussian Naive Bayes.

---

# 4. Gaussian Naive Bayes With Synthetic Minority Oversampling Technique Using RobustScaler

Using RobustScaler instead of StandardScaler produced very similar behavior.

The model achieved:

* Accuracy score = `47.76%`
* Minority recall = `0.74`
* Minority F1-score = `0.37`
* Receiver Operating Characteristic - Area Under Curve = `0.576`

Although RobustScaler slightly reduced the minority recall compared to StandardScaler, the overall performance remained highly unstable.

The model still generated a large number of false positives and struggled to maintain balanced predictive capability.

This experiment suggests that changing the scaling strategy alone was insufficient to stabilize Gaussian Naive Bayes performance after applying Synthetic Minority Oversampling Technique.

---

# 5. Gaussian Naive Bayes With Synthetic Minority Oversampling Technique, RobustScaler, and One-Hot Encoding With Drop First

This experiment produced the strongest overall performance among all Gaussian Naive Bayes configurations.

The model achieved:

* Accuracy score = `71.60%`
* Minority recall = `0.71`
* Minority F1-score = `0.50`
* Receiver Operating Characteristic - Area Under Curve = `0.713`

Compared to previous experiments, the addition of one-hot encoding with dropping the first category significantly improved model stability and overall classification balance.

The confusion matrix showed a substantial reduction in false positives while maintaining strong minority-class recall.

This experiment demonstrates that feature representation and encoding strategies strongly influence Gaussian Naive Bayes performance.

---

# 6. Gaussian Naive Bayes With Synthetic Minority Oversampling Technique, StandardScaler, and One-Hot Encoding With Drop First

This experiment also produced strong and balanced performance.

The model achieved:

* Accuracy score = `71.56%`
* Minority recall = `0.69`
* Minority F1-score = `0.50`
* Receiver Operating Characteristic - Area Under Curve = `0.707`

The results remained extremely close to the RobustScaler configuration.

This experiment confirms that combining:

* Synthetic Minority Oversampling Technique
* Feature scaling
* One-hot encoding optimization

significantly improves Gaussian Naive Bayes performance in imbalanced classification problems.

# Key Insights

## Accuracy Alone Is Misleading in Imbalanced Classification Problems

The baseline model achieved high accuracy while completely failing to detect minority-class samples. This confirms that accuracy alone is not a sufficient metric for evaluating imbalanced datasets.

---

## Gaussian Naive Bayes Is Highly Sensitive to Imbalance

Without imbalance-handling strategies, Gaussian Naive Bayes strongly favored the majority class and ignored minority samples entirely.

This demonstrates the importance of preprocessing and balancing techniques when using probabilistic classifiers.

---

## Prior Probability Adjustment Improved Minority-Class Detection

Balancing class priors significantly improved minority recall and Receiver Operating Characteristic - Area Under Curve performance.

This shows that prior probability tuning can partially compensate for class imbalance in Gaussian Naive Bayes models.

---

## Synthetic Minority Oversampling Technique Increased Recall but Introduced False Positives

Applying Synthetic Minority Oversampling Technique greatly improved minority recall but also generated excessive false-positive predictions in several experiments.

This highlights the trade-off between sensitivity and precision in imbalanced classification problems.

---

## One-Hot Encoding Optimization Significantly Improved Performance

The strongest results were achieved after combining:

* Synthetic Minority Oversampling Technique
* Feature scaling
* One-hot encoding with dropping the first category

This indicates that feature representation quality strongly affects Gaussian Naive Bayes performance.

---

## RobustScaler and StandardScaler Produced Similar Final Performance

Both scaling strategies achieved highly competitive results when combined with optimized preprocessing pipelines.

This suggests that preprocessing consistency was more important than the specific scaling method itself.

---

# Conclusion

This experimentation demonstrated that Gaussian Naive Bayes performance is strongly influenced by imbalance-handling techniques, preprocessing strategies, and feature representation methods.

The experiments showed that:

* Gaussian Naive Bayes fails to detect minority-class samples without imbalance handling
* Prior probability adjustment improves recall performance
* Synthetic Minority Oversampling Technique increases minority sensitivity but may introduce false positives
* One-hot encoding optimization significantly improves model stability and predictive balance
* Combining preprocessing strategies produces substantially stronger classification performance

The strongest overall configuration was Gaussian Naive Bayes with Synthetic Minority Oversampling Technique, RobustScaler, and one-hot encoding with dropping the first category, as it achieved the best balance between minority recall, F1-score, Receiver Operating Characteristic - Area Under Curve, and overall predictive stability.

These experiments highlight the importance of systematic preprocessing experimentation and careful evaluation when building probabilistic machine learning pipelines for imbalanced datasets.
