# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

# **Outlier Analysis**
Outlier detection was performed using the IQR method on the numerical features.

### Features with Detected Outliers

- `CreditScore`
- `Age`
- `NumOfProducts`
- `Exited`

### Decision About Outliers

The detected outliers were intentionally retained instead of being removed because these values represent realistic customer behavior rather than data entry errors.

#### CreditScore
Extreme credit scores are valid in banking datasets and may contain important information about customer financial behavior and churn patterns. Removing these observations could reduce the model’s ability to learn from high-risk or low-risk customers.

#### Age
Older customers were detected as statistical outliers due to the IQR method. However, these customers are valid observations and may have significantly different churn behavior compared to younger customers. Therefore, they were preserved.

#### NumOfProducts
Customers with a high number of banking products were identified as outliers. These cases are meaningful from a business perspective because product usage can strongly influence customer retention and churn.

#### Exited
The `Exited` feature represents the target variable of the classification problem, where:

- `0` = Customer stayed
- `1` = Customer churned

Although the IQR method detected many values as outliers, this result is not meaningful because `Exited` is a binary categorical variable rather than a continuous numerical feature.

Outlier detection methods such as IQR are designed for continuous numerical distributions and should not be applied to classification labels.

Therefore, no outlier handling was performed on the `Exited` column, and all observations were retained.

### Additional Notes
- Identifier columns such as:
  - `RowNumber`
  - `CustomerId`

  were removed from the dataset because they do not provide predictive value.

- `Surname`

The `Surname` column was removed from the dataset because it represents a high cardinality categorical feature with a large number of unique customer names.

Customer surnames are unlikely to have a direct causal relationship with churn behavior and may introduce noise into the model.

Additionally, encoding this feature could create unnecessary dimensionality or artificial relationships between categories without providing meaningful predictive value.

Therefore, the `Surname` feature was excluded from the modeling process.

### Conclusion

The outliers were retained to preserve potentially valuable business information and avoid removing legitimate customer behavior patterns. Tree-based models such as Random Forest are generally robust to outliers, making aggressive outlier removal unnecessary for this project.

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False, drop='first'),[1,2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
```

### Categorical Feature Encoding

Categorical variables were transformed into numerical representations using One-Hot Encoding.

#### Geography
The `Geography` feature was encoded using One-Hot Encoding. Since the feature contains three categories (`France`, `Spain`, and `Germany`), two binary columns were created while one category was automatically dropped to avoid the dummy variable trap and multicollinearity.

Encoding representation:

| Germany | Spain | Geography |
|---|---|---|
| 0 | 0 | France |
| 0 | 1 | Spain |
| 1 | 0 | Germany |

#### Gender
The `Gender` feature was also transformed using One-Hot Encoding. Since it contains only two categories (`Female` and `Male`), a single binary column was generated after dropping one category.

Encoding representation:

| Gender |
|---|
| 0 → Female |
| 1 → Male |

These transformations convert categorical variables into numerical representations suitable for machine learning algorithms.

# **Building Support Vector Machine (SVM) Classification Model**

## Training the Building Support Vector Machine (SVM) Classification model on the Training set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', class_weight = 'balanced', random_state = 0)
classifier.fit(X_train, y_train)
```

# **Handling Class Imbalance using `class_weight = 'balanced'`**

# **Preticting the Test set results**

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)[0:100,:])
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)

print("\nMean Accuracy:")
print(accuracies.mean())

print("\nStandard Deviation:")
print(accuracies.std())
```

# Support Vector Classification Model Comparison and Evaluation

This experiment focused on evaluating multiple configurations of Support Vector Machine models for an imbalanced classification problem. The comparison analyzed the impact of:

* Different kernel functions
* Class imbalance handling using `class_weight = 'balanced'`
* Feature scaling strategies
* Minority-class detection capability
* Model stability and generalization performance

The experiments aimed to determine which Support Vector Classification configuration provides the strongest balance between overall predictive performance and minority-class detection.

---

# 1. Linear Kernel Without Handling Imbalance Using StandardScaler

The linear kernel model without imbalance handling produced the weakest practical classification performance despite achieving an accuracy score of `79.64%`.

The confusion matrix revealed that the model predicted every sample as the majority class and completely failed to detect any minority-class samples. As a result:

* Minority-class precision = `0.00`
* Minority-class recall = `0.00`
* Minority-class F1-score = `0.00`
* Receiver Operating Characteristic - Area Under Curve = `0.50`

A Receiver Operating Characteristic - Area Under Curve score of `0.50` indicates that the model performed no better than random guessing for class separation.

This experiment clearly demonstrates the danger of relying solely on accuracy in imbalanced classification problems. Although the model achieved relatively high accuracy, it entirely ignored the minority class.

The warning messages generated during evaluation further confirmed that the model never predicted minority-class samples.

---

# 2. Linear Kernel With Balanced Class Weights Using StandardScaler

After introducing balanced class weights, the linear kernel model improved significantly in minority-class detection.

The minority-class recall increased dramatically from `0.00` to `0.73`, showing that the model became capable of identifying most minority-class samples. The Receiver Operating Characteristic - Area Under Curve score also improved substantially to `0.721`.

However, this improvement came with a decrease in overall accuracy to `71.36%`. This behavior is expected because the model became less biased toward the majority class.

The minority-class precision remained relatively low at `0.39`, meaning the model generated a considerable number of false positives.

This experiment demonstrates the effectiveness of imbalance handling techniques in improving minority-class detection within Support Vector Classification models.

---

# 3. Radial Basis Function Kernel With Balanced Class Weights Using StandardScaler

The Radial Basis Function kernel model produced the strongest overall performance among all experiments.

The model achieved:

* Accuracy score = `78.92%`
* Minority-class recall = `0.76`
* Minority-class F1-score = `0.60`
* Receiver Operating Characteristic - Area Under Curve = `0.780`

The Radial Basis Function kernel provided a strong balance between:

* Overall accuracy
* Minority-class detection
* Precision
* Generalization capability

Compared to the linear kernel, the Radial Basis Function kernel captured more complex non-linear decision boundaries in the dataset, leading to improved classification performance.

This experiment demonstrates that non-linear kernels can significantly improve classification performance in complex datasets.

---

# 4. Sigmoid Kernel With Balanced Class Weights Using StandardScaler

The sigmoid kernel produced the weakest overall performance among the balanced models.

The model achieved:

* Accuracy score = `53.04%`
* Minority-class F1-score = `0.33`
* Receiver Operating Characteristic - Area Under Curve = `0.542`

Although the recall score for the minority class reached `0.56`, the model generated an extremely large number of false positives, as shown by the high number of incorrectly classified majority-class samples.

The poor Receiver Operating Characteristic - Area Under Curve score indicates weak class separation capability.

This experiment demonstrates that the sigmoid kernel was not suitable for the dataset characteristics.

---

# 5. Polynomial Kernel With Balanced Class Weights Using StandardScaler

The polynomial kernel achieved strong and balanced performance.

The model achieved:

* Accuracy score = `77.48%`
* Minority-class recall = `0.76`
* Minority-class F1-score = `0.58`
* Receiver Operating Characteristic - Area Under Curve = `0.771`

The polynomial kernel performed slightly below the Radial Basis Function kernel but still demonstrated strong classification capability.

The results indicate that the polynomial kernel successfully captured non-linear patterns in the data while maintaining relatively stable overall performance.

This experiment confirms that non-linear Support Vector Classification kernels outperform linear kernels for this dataset.

---

# 6. Radial Basis Function Kernel With Balanced Class Weights Using RobustScaler

Using RobustScaler with the Radial Basis Function kernel produced highly competitive results.

The model achieved:

* Accuracy score = `78.76%`
* Minority-class recall = `0.77`
* Minority-class F1-score = `0.59`
* Receiver Operating Characteristic - Area Under Curve = `0.780`

The performance remained extremely close to the StandardScaler experiment, indicating that the Radial Basis Function kernel remained stable across different scaling strategies.

This experiment also demonstrates that RobustScaler can effectively handle datasets containing outliers or skewed numerical distributions.

Although the StandardScaler configuration slightly outperformed RobustScaler in overall balance, both configurations produced strong results.
