# **Data preprocessing**

## Data Cleaning Check Template
This template is designed to quickly assess the quality of any dataset before building machine learning models or performing analysis.

It provides a structured overview of the dataset by checking for common data issues such as:

- Missing values

- Duplicate rows

- Incorrect data types

- Outliers

- Distribution of numerical features

- Categorical feature consistency

**What This Template Does**

- Displays basic dataset information (shape, data types)

- Identifies missing values and duplicates

- Summarizes numerical and categorical features

- Detects potential outliers using the IQR method

- Highlights columns with low unique values for quick inspection

How to Use

1. Load your dataset using Pandas  

2. Call the function:

```python
def data_quality_report(df):

    print("DATA QUALITY REPORT")
    
    # Print a separator line for better readability
    
    print("=" * 50)
    print("BASIC INFO")
    print("=" * 50)
    
    # Show general information about the dataset (columns, data types, non-null values)
    print(df.info())
    
    # Show number of rows and columns
    print("\n" + "=" * 50)
    print("SHAPE OF DATA")
    print("=" * 50)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    # Check for missing (null) values in each column
    print("\n" + "=" * 50)
    print("MISSING VALUES")
    print("=" * 50)
    missing = df.isnull().sum()
    
    # Display only columns that have missing values
    print(missing[missing > 0])
    
    # Check for duplicate rows
    print("\n" + "=" * 50)
    print("DUPLICATES")
    print("=" * 50)
    print(f"Duplicate rows: {df.duplicated().sum()}")
    
    # Display data types of each column
    print("\n" + "=" * 50)
    print("DATA TYPES")
    print("=" * 50)
    print(df.dtypes)
    
    # Summary statistics for numerical columns (mean, std, min, max, etc.)
    print("\n" + "=" * 50)
    print("NUMERICAL SUMMARY")
    print("=" * 50)
    print(df.describe())
    
    # Summary for categorical (object) columns
    print("\n" + "=" * 50)
    print("CATEGORICAL SUMMARY")
    print("=" * 50)
    print(df.describe(include=['object']))
    
    # Show unique values for columns with low number of distinct values
    # Useful for detecting categories, errors, or inconsistencies
    print("\n" + "=" * 50)
    print("UNIQUE VALUES (LOW CARDINALITY)")
    print("=" * 50)
    for col in df.columns:
        if df[col].nunique() < 10:  # Only show columns with few unique values
            print(f"{col}: {df[col].unique()}")
            
    # correlation
    print("\n" + "=" * 50)
    print("CORRELATION MATRIX")
    print("=" * 50)
    print(df.corr(numeric_only=True))
    
    # Detect outliers using the IQR (Interquartile Range) method
    print("\n" + "=" * 50)
    print("OUTLIERS CHECK (IQR METHOD)")
    print("=" * 50)
    
    # Loop through only numerical columns
    for col in df.select_dtypes(include=np.number).columns:
        Q1 = df[col].quantile(0.25)  # 25th percentile
        Q3 = df[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range
        
        # Count rows that fall outside the normal range
        outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)]
        print(f"{col}: {len(outliers)} outliers")
```

### Decision About Outliers

The detected outliers were intentionally retained instead of being removed because these values represent realistic customer behavior rather than data entry errors.

#### CreditScore
Extreme credit scores are valid in banking datasets and may contain important information about customer financial behavior and churn patterns. Removing these observations could reduce the model’s ability to learn from high-risk or low-risk customers.

#### Age
Older customers were detected as statistical outliers due to the IQR method. However, these customers are valid observations and may have significantly different churn behavior compared to younger customers. Therefore, they were preserved.

#### NumOfProducts
Customers with a high number of banking products were identified as outliers. These cases are meaningful from a business perspective because product usage can strongly influence customer retention and churn.

#### Exited
The `Exited` feature represents the target variable of the classification problem, where:

- `0` = Customer stayed
- `1` = Customer churned

Although the IQR method detected many values as outliers, this result is not meaningful because `Exited` is a binary categorical variable rather than a continuous numerical feature.

Outlier detection methods such as IQR are designed for continuous numerical distributions and should not be applied to classification labels.

Therefore, no outlier handling was performed on the `Exited` column, and all observations were retained.

### Additional Notes
- Identifier columns such as:
  - `RowNumber`
  - `CustomerId`

  were removed from the dataset because they do not provide predictive value.

- `Surname`

The `Surname` column was removed from the dataset because it represents a high cardinality categorical feature with a large number of unique customer names.

Customer surnames are unlikely to have a direct causal relationship with churn behavior and may introduce noise into the model.

Additionally, encoding this feature could create unnecessary dimensionality or artificial relationships between categories without providing meaningful predictive value.

Therefore, the `Surname` feature was excluded from the modeling process.

### Conclusion

The outliers were retained to preserve potentially valuable business information and avoid removing legitimate customer behavior patterns. Tree-based models such as Random Forest are generally robust to outliers, making aggressive outlier removal unnecessary for this project.

# **Encoding categorical data**

## Encoding the Indpendent Variable Using One Hot Encoding

Note: I explained One-Hot Encoding earlier.

If you missed it or want a deeper understanding of how it works, you can find it here: [Go to One-Hot Encoding Section](https://github.com/Hazem1695/ml-concept-briefs)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'color' column (index 2) while keeping other features
# Initialize ColumnTransformer to handle categorical data
# [2] targets the 'color' column (index 2)
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(sparse_output=False, drop='first'),[1,2])], remainder = 'passthrough')
# Execute transformation and convert to NumPy ndarray for model training
X = np.array(ct.fit_transform(X))
```

# **Building Logistic Regression Classification Model**

## Training the Building Logistic Regression Classification model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(class_weight = 'balanced', random_state = 0)
classifier.fit(X_train,y_train)
```

### Handling Class Imbalance

The dataset contains an imbalanced target distribution, where the number of non-churned customers is significantly higher than churned customers.

To reduce bias toward the majority class, the `class_weight='balanced'` parameter was applied during model training. This automatically adjusts class weights inversely proportional to class frequencies, giving more importance to minority class samples (`Exited = 1`).

This approach helps improve the model’s ability to detect churned customers and increases recall for the minority class.

---

### Difference Between Handling Class Imbalance Methods

| ------------- | ---------------------------------- |
| Class Weights | changes learning importance        |
| SMOTE         | creates synthetic minority samples |
| Undersampling | removes majority samples           |


# **Preticting the Test set results**

```python
y_pred = classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)[0:100,:])
```

# **Evaluating the Model Performance**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_auc_score
from sklearn.model_selection import cross_val_score

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nROC-AUC Score:")
print(roc_auc_score(y_test, y_pred))


accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
print("\nMean Accuracy:")
print(accuracies.mean())
print("\nStandard Deviation:")
print(accuracies.std())

```

# Model Comparison Analysis

This experiment evaluates the impact of handling class imbalance using `class_weight='balanced'` and compares the effect of two different feature scaling techniques: `StandardScaler` and `RobustScaler`.

The main objective was not only to maximize overall accuracy, but also to improve the model’s ability to correctly identify the minority class, which is the most important challenge in imbalanced classification problems such as customer churn prediction.

---

# 1. Baseline Model

## Without `class_weight='balanced'` using `StandardScaler`

The baseline model achieved the highest overall accuracy score at **80.68%**, which may initially appear to indicate strong performance. However, deeper evaluation reveals that the model struggled significantly with detecting the minority class.

The confusion matrix shows that the model correctly identified most majority-class samples, achieving a very high recall of **95%** for class `0`. However, it only detected **116** out of **509** minority-class samples, resulting in a minority-class recall of just **23%**.

This behavior demonstrates a common issue in imbalanced datasets: the model becomes biased toward the dominant class and achieves high accuracy simply by predicting the majority class more frequently.

Although the weighted metrics appear acceptable, the minority-class performance remained weak:

* Minority-class precision: **56%**
* Minority-class recall: **23%**
* Minority-class F1-score: **32%**
* Receiver Operating Characteristic - Area Under Curve score: **0.591**

The low Receiver Operating Characteristic - Area Under Curve score indicates limited class separation capability and poor balance between true positive and false positive predictions.

---

# 2. Balanced Model

## With `class_weight='balanced'` using `StandardScaler`

After introducing `class_weight='balanced'`, the model became significantly more effective at identifying minority-class samples.

The minority-class recall increased dramatically from **23%** to **73%**, meaning the model successfully captured a much larger portion of positive cases. This improvement is highly important in classification tasks where missing minority-class samples carries a high business cost.

The confusion matrix confirms this improvement:

* False negatives decreased from **393** to **136**
* True positives increased from **116** to **373**

Although overall accuracy dropped from **80.68%** to **70.32%**, this trade-off is expected in imbalanced classification tasks because the model is no longer overly biased toward the majority class.

Additional improvements include:

* Receiver Operating Characteristic - Area Under Curve score improved from **0.591** to **0.714**
* Macro-average recall improved substantially
* Minority-class F1-score increased from **32%** to **50%**

This model demonstrates a much more balanced learning behavior and provides stronger generalization for real-world imbalanced classification scenarios.

---

# 3. Balanced Model

## With `class_weight='balanced'` using `RobustScaler`

The model trained with `RobustScaler` produced nearly identical results to the model trained with `StandardScaler`.

Performance differences were extremely small:

* Accuracy improved slightly from **70.32%** to **70.36%**
* Receiver Operating Characteristic - Area Under Curve score improved marginally from **0.7142** to **0.7145**
* Cross-validation mean accuracy improved slightly
* Standard deviation increased marginally

The confusion matrix and classification report remained effectively unchanged.

This indicates that the dataset was not heavily affected by extreme outliers, since `RobustScaler` is specifically designed to reduce the influence of outlier values. As a result, replacing `StandardScaler` with `RobustScaler` did not lead to meaningful performance improvements.

# Insights

## 1. Accuracy Alone Was Misleading

The baseline model achieved the highest accuracy, but failed to identify most minority-class samples. This demonstrates why accuracy should not be the primary evaluation metric for imbalanced classification problems.

---

## 2. Balanced Class Weight Significantly Improved Minority Detection

Applying `class_weight='balanced'` transformed the model behavior:

* Minority recall increased from **23%** to **73%**
* False negatives were reduced substantially
* Receiver Operating Characteristic - Area Under Curve score improved considerably

This indicates that class balancing helped the model learn decision boundaries that better represent both classes instead of favoring the majority class.

---

## 3. The Model Became More Balanced Rather Than More Accurate

After balancing, the model sacrificed some majority-class accuracy in exchange for dramatically better minority detection.

This trade-off is often desirable in real-world applications such as:

* Customer churn prediction
* Fraud detection
* Medical diagnosis
* Risk prediction systems

where missing positive cases can be more costly than generating additional false positives.

---

## 4. RobustScaler Did Not Provide Significant Improvement

The nearly identical performance between `StandardScaler` and `RobustScaler` suggests that:

* Outliers were not a major issue in the dataset
* The model performance was influenced more by class imbalance than scaling strategy

This experiment highlights the importance of understanding the actual bottleneck affecting model performance instead of applying preprocessing techniques blindly.

---

# Conclusion

The final selected configuration was:

## `class_weight='balanced'` with `StandardScaler`

Although this configuration achieved lower overall accuracy compared to the baseline model, it provided substantially better minority-class detection and a much stronger Receiver Operating Characteristic - Area Under Curve score.

The experiment demonstrates an important machine learning principle:

> A model with lower accuracy can still be significantly better if it improves the ability to correctly classify the minority class.

From a practical machine learning perspective, the balanced model is more reliable and more suitable for real-world imbalanced classification tasks because it prioritizes meaningful class detection rather than inflated accuracy scores driven by majority-class dominance.
